```python
#linear regression of area and perimeter
```

```python
area = data['Area']
area
```

```python
perm = data['Perimeter']
perm
```

```python
#create test model and values
```

```python
from sklearn.model_selection import train_test_split

train_X, test_X, train_y, test_y = train_test_split(area, perm, random_state=42)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

ax.scatter(train_X, train_y, color="red", marker="o", label="train")
ax.scatter(test_X, test_y, color="blue", marker="x", label="test")
ax.legend()
```

```python
#reshape data into numpy array
```

```python
import numpy as np

train_X = np.array(train_X).reshape(-1, 1)
test_X = np.array(test_X).reshape(-1, 1)
```

```python
#fit linear regression
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression(fit_intercept=True)
model.fit(train_X, train_y)
```

```python
train_score = model.score(train_X, train_y)
print("The training score of the model is: ", train_score)
test_score = model.score(test_X, test_y)
print("The score of the model on test data is:", test_score )
```

```python
#model score is very close to 1, therefore the model is a good model
```

```python
#visualise linear regression
```

```python
import matplotlib.pyplot as plt

y_pred = model.predict(test_X)
plt.scatter(test_X, test_y, color='b')
plt.xlabel('Area')
plt.ylabel('Perimeter')
plt.title('Linear regression to fit Test Data')
plt.plot(test_X, y_pred, color='k')
plt.show()
```

```python
#graph shows clear linear regression
```

```python
#copy dataset to carry out cluster analysis
```

```python
import copy
```

```python
data2 = copy.deepcopy(data)
```

```python
data2.head()
```

```python
#K means cluster analysis
```

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
X = data2
Y = data2
Z = np.vstack((X,Y))
```

```python
#convert to numpy float
```

```python
Z = np.float32(Z)
```

```python
#define criteria and apply kmeans analysis
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
ret,label,center = cv2.kmeans(Z,2,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
```

```python
#seperate the data
```

```python
A = Z[label.ravel()==0]
B = Z[label.ravel()==1]
```

```python
#plot the data
```

```python
plt.scatter(A[:,0],A[:,1])
plt.scatter(B[:,0],B[:,1],c = 'r')
plt.scatter(center[:,0],center[:,1],s = 80,c = 'y', marker = 's')
plt.xlabel('Test Data'),plt.ylabel('Z samples')
plt.show()
```

```python
#shows two clusters, but the edge of the clusters are slightly overlapping
```

```python
#however there are three varities or wheat, not two, therefore each variety does not belong to a seperate cluster
```
